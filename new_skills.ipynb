### 'School Catalalogue' project
In this project I was tasked with creating a digital catalogue for the some schools to hold quick reference info for each school in the city. I needed to create classes for primary and secondary schools but because the two share properties and methods, I had them inherit from a parent 'school' class.

```python
class School:
    def __init__(self, name, level, numberOfStudents):
        self.name = name
        self.level = level
        self.numberOfStudents = numberOfStudents

    def get_name(self):
        return self.name

    def get_level(self):
        return self.level

    def get_numberOfStudents(self):
        return self.numberOfStudents

    def set_numberOfStudents(self, num):
        self.numberOfStudents = num

    def __repr__(self):
        txt = 'A {level} school named {name} with {numberOfStudents} students'.format(level = self.level, name = self.name, numberOfStudents = self.numberOfStudents)
        return txt

class PrimarySchool(School):
    def __init__(self, name, numberOfStudents, pickupPolicy):
        super().__init__(name, 'primary', numberOfStudents)
        self.pickupPolicy = pickupPolicy

    def get_pickupPolicy(self):
        return self.pickupPolicy

    def __repr__(self):
        parentRepr = super().__repr__()
        return parentRepr + '. The pickup policy is {pickupPolicy}'.format(pickupPolicy = self.pickupPolicy)

class HighSchool(School):
    def __init__(self, name, numberOfStudents, sportsTeams):
        super().__init__(name, 'high', numberOfStudents)
        self.sportsTeams = sportsTeams

    def get_sportsTeams(self):
        return self.sportsTeams

    def __repr__(self):
        txt = 'Sports teams include: {sportsTeams}'.format(sportsTeams = self.sportsTeams)
        return txt

# Testing:
#a1 = School('Rivington', 'Primary', 500)
#a1.set_numberOfStudents(404)
#a1.get_numberOfStudents
#print(a1)

# Testing:
#b1 = PrimarySchool('Rivington', 709, 'pickup from Dad')
#print(b1)
#print(b1.get_level())

# Testing:
#c1 = HighSchool('Rainford', 386, ['Athletics', 'Football', 'Tennis'])
#print(c1.get_name())
#print(c1.get_level())
#print(c1.get_sportsTeams())
#print(c1)
```
